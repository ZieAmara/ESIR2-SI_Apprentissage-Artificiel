## <font color="#9400D3">2. Fitting a decision tree</font>

The basic decision tree can be obtained easily using the class <code>DecisionTreeClassifier</code>. More advanced options can be inserted into the 'DecisionTreeClassifier' function.

Remember:  The <code>fit</code> function needs to have 2 parameters (at least) : 
- the set of features describing the examples (here the 2 first columns of our train set)

```python
# instantiate the decision tree model (using most of the default parameters)
dt = tree.DecisionTreeClassifier(criterion='entropy').fit(data_train.iloc[:,:2], data_train.Y)
```

### Decision boundaries

The function <code>draw_boundary_tree</code> plots the decision boundary (only for dataset with 2 features of course).
It takes as parameters : 
 - the model (here a tree)
 - the dataset of examples that you want to plot together with the boundary
An example of the use of this function with our data and tree is given right after

```python
def draw_boundary_tree(model, data):
    h = 0.05
    xx, yy = np.meshgrid(
            np.linspace(data.iloc[:, 0].min(), data.iloc[:, 0].max()),
            np.linspace(data.iloc[:, 1].min(), data.iloc[:, 1].max()))
    zz = np.c_[xx.ravel(), yy.ravel()]
    zz = pd.DataFrame(zz)
    zz2 = zz
    zz2.columns=['X1','X2'] # avoid warning on valid feature names
    pred_zz= pd.Series(model.predict(zz2))
    color_map = matplotlib.colors.ListedColormap(pd.Series(['blue', 'red']))
    fig = plt.figure(figsize=  (8,8))
    fig = plt.scatter(zz.iloc[:,0], zz.iloc[:,1], c = pred_zz, cmap = color_map, marker='+', s=70)
    fig = plt.scatter(data.iloc[:,0], data.iloc[:,1], s = 50, c = data.iloc[:,2], cmap = color_map)
```

```python
draw_boundary_tree(dt, data_train)
```

<font color="red">**Question 6:**</font> Why this boundary is typical to a decision tree ? What kind of phenomenon can you observe here ?

Answer: The decision region is defined by multiple straight horizontal and vertical lines, corresponding to binary tests on attributes, resulting in splits orthogonal to the attribute dimension.

<font color="brown">--> **interprétation :**</font> 
    
    Cette limite est typique à arbre de décision car tout comme l'arbre de décision, elle divise les données en groupes en utilisant une logique simple de sélection basée sur la valeur de la variable la plus discriminante.

    Le graphique ci-dessus montre les limites de prise de décision du modèle. Les points bleus et rouges représentent les données d'entraînement. Les points Bleu en zone Bleu et point rouge en zone rouge sont correctement classés par le modèle, tandis que les autres points sont mal classés.
    L'arbre de décision a subi un phénomène de surapprentissage en produisant une frontière qui suit de près les données d'entraînement. Cela peut entraîner une mauvaise performance pour les données de test car il ne peut pas généraliser les résultats pour les nouvelles données

## <font color="#9400D3">3. Pruning the tree</font>

The 'basic' decision tree $T_{max}$ is only composed of pure leaves which might lead to an overfitting model. 

There are 2 ways to simplify the model (and thus increase his generalization ability):
- **pre-pruning**: early stop of the tree building. Can be controlled with parameters 'max_depth', 'min_samples_split', 'min_samples_leaf', 'max_leaf_nodes' or 'min_impurity_decrease'
- **post-pruning** with a validation set.

In the second case, a common algorithm is the **minimal cost-complexity pruning**. This algorithm is parameterized by $\alpha \geq 0$, known as the complexity parameter. A non-terminal node with the smallest value of $\alpha$ is the weakest link and will be pruned.

### Effective alphas of pruned tree

<code>cost_complexity_pruning_path</code> returns the $\alpha$ and the corresponding total leaf impurities at each step of the pruning process. As $\alpha$ increases, more of the tree is pruned, which increases the total impurity of its leaves.


```python
tr = tree.DecisionTreeClassifier(criterion='entropy') # first declare a tree
path = tr.cost_complexity_pruning_path(data_train.iloc[:,0:2], data_train.Y) # then ask for all the possible values 
# of alpha to prune this tree (depends on the training set)
print(path)
```

You should see above that, for our training set, different trees can be obtained, from basic tree ($\alpha = 0 \rightarrow$ no pruning, no impurity) to maximum pruning.

The values above correspond to the different values of $\alpha$ that we can use ('ccp_alpha') and the associated average impurity in the leaves ('impurities').

We plot below the level of impurity vs the values of $\alpha$. 

```python
ccp_alphas, impurities = path.ccp_alphas, path.impurities
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, impurities)
plt.xlabel("effective alpha")
plt.ylabel("total impurity of leaves")
```

If you want to build a tree with a particular value of $\alpha$, you need to specify it into the 'DecisionTreeClassifier' function (see below)

```python
# Exemple of a pruned tree with the third value from all the possible alphas (the first one is 0, already done with the basic tree)
tr_pruned = tree.DecisionTreeClassifier(criterion='entropy',ccp_alpha=path.ccp_alphas[2]).fit(data_train.iloc[:,0:2], data_train.Y)
```

```python
# TODO

# Draw the pruned tree
plt.figure(figsize=(10, 6))
tree.plot_tree(tr_pruned, filled=True)
plt.show()

# Draw the boundary of the pruned tree
draw_boundary_tree(tr_pruned, data_train)

# Compute the performance on the training set
print("La performance sur le plateau d'entraînement est ", tr_pruned.score(data_train.iloc[:,:2], data_train.Y))

# Compute the performance on the validation set
print("La performance sur l'ensemble de validation est ", tr_pruned.score(data_valid.iloc[:,:2], data_valid.Y))
```

### Pruned tree selection

<font color="blue">**todo:**</font> Use the validation set to select the best pruned tree :
 - Use a for loop on the values of alpha (path.ccp_alphas) to compute the prediction scores (accuracy) of all the possible pruned trees on the training set and on the validation set (you can store the scores in 2 different vectors : 'train_score' and 'valid_score').
 - Plot on a graph these scores VS alpha (2 different curves). 


```python
# TO COMPLETE

train_scores, valid_scores = np.zeros(len(path.ccp_alphas)), np.zeros(len(path.ccp_alphas))

for i in range(len(path.ccp_alphas)):
    tr_pruned = tree.DecisionTreeClassifier(criterion='entropy',ccp_alpha=path.ccp_alphas[i]).fit(data_train.iloc[:,0:2], data_train.Y)
    train_scores[i] = tr_pruned.score(data_train.iloc[:,:2], data_train.Y)
    valid_scores[i] = tr_pruned.score(data_valid.iloc[:,:2], data_valid.Y)

print('train_scores =>', train_scores, '\n\nvalid_scores =>', valid_scores)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and validation sets")
ax.plot(path.ccp_alphas, train_scores, marker="o", label="train", drawstyle="steps-post")
ax.plot(path.ccp_alphas, valid_scores, marker="o", label="valid", drawstyle="steps-post")
ax.legend()
plt.show()
```

<font color="red">**Question 7:**</font>

 - Which pruned tree will you choose here ? What are its number of nodes and leaves ? Draw its decision boundary
 - Estimate its generalization error
 - Compare this error with the $T_{max}$ tree, as well as the complexity (number of nodes/leaves). Conclude. 

```python
# TODO 

# which pruned tree is the best?
# The best pruned tree is the one with the highest accuracy on the validation set.
best_tr_alpha = path.ccp_alphas[np.argmax(valid_scores)]
print("Le meilleur alpha est",best_tr_alpha)
best_tr = tree.DecisionTreeClassifier(criterion='entropy', ccp_alpha=best_tr_alpha).fit(data_train.iloc[:,:2], data_train.Y)

# Compute the number of nodes and leaves of the best pruned tree
print("\nLe meilleur arbre élagué a ",best_tr.tree_.node_count,"Nœuds et",best_tr.tree_.n_leaves,"Feuilles.")

# Estimate the generalization error of the best pruned tree on the test set
print("\nErreur de généralisation de l'arbre élagué : ",1 - best_tr.score(data_test.iloc[:,:2], data_test.Y))

# Compare the generalization error of the best pruned tree with the generalization error of the basic tree
tr = tree.DecisionTreeClassifier(criterion='entropy').fit(data_train.iloc[:,0:2], data_train.Y)
print("Erreur de généralisation de l'arbre de base : ",1 - tr.score(data_test.iloc[:,:2], data_test.Y))

plt.figure(figsize=(30, 20))
tree.plot_tree(best_tr, filled=True)
plt.show()

# Compare the complexity of the best pruned tree with the complexity of the basic tree
print("\n\tLa complexité du meilleur arbre élagué est",best_tr.tree_.node_count,"\n\tEt celui de l'arbre de base est : ",tr.tree_.node_count)

# Same for Leaves
print("\n\tLe nombre de feuilles du meilleur arbre élagué est",best_tr.tree_.n_leaves,"\n\tEt celui de l'arbre de base est : ",tr.tree_.n_leaves)

# same for nodes
print("\n\tLe nombre de nœuds du meilleur arbre élagué est",best_tr.tree_.node_count,"\n\tEt celui de l'arbre de base est : ",tr.tree_.node_count)

```

<font color="brown">--> **conclusion :**</font>

    Au vu des résultats obtenu, nous savons que l'arbre choisi a moins de feuilles et des noeuds moindre comparé à ceux de l'arbre maximum; il est donc plus simple. cependant son erreur de génération est bien plus élevée que celle de l'arbre maximum. On peut donc conclure que même si l'arbre réduit est moins complexe que l'arbre $T_{max}$ en termes de nombre de feuilles et de nœuds, il n'est pas aussi performant en termes d'erreur de généralisation.  l'arbre $T_{max}$ a une meilleure capacité à généraliser les résultats pour des données non vues. Il est donc préferable de choisir l'arbre l'arbre $T_{max}$ au lieu de l'arbre réduit.

```python
# Draw of decision boundary
draw_boundary_tree(best_tr, data_train)
```

```python
# TODO
# Construire un arbre élagué en utilisant un critère d'arrêt précoce

## pré-élagage avec max-daph
tr_pruned_pre = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4).fit(data_train.iloc[:,0:2], data_train.Y)

## post-élagage
# construction d'un arbre de décision complet
tr_pruned_post = tree.DecisionTreeClassifier(criterion='entropy').fit(data_train.iloc[:,0:2], data_train.Y)
# évaluation de la performance de l'arbre de décision complet sur l'ensemble de validation
val_score = tr_pruned_post.score(data_valid.iloc[:,:2], data_valid.Y)
print("la performance sur le jeu de validation pour l'arbre complet :", val_score)
# post-élagage avec ensemble de validation
best_alpha = None
best_score = 0
for i in range (0,len(valid_scores)-1):
    dtc = tree.DecisionTreeClassifier(criterion='entropy', ccp_alpha=path.ccp_alphas[i]).fit(data_train.iloc[:,:2], data_train.Y)
    score = dtc.score(data_valid.iloc[:,:2], data_valid.Y)
    if score > best_score:
        best_score = score
        best_alpha = path.ccp_alphas[i]    
print("la meilleure valeur de alpha par post-élagage est : ", best_alpha)
tr_pruned_post = tree.DecisionTreeClassifier(criterion='entropy', ccp_alpha=best_alpha).fit(data_train.iloc[:,0:2], data_train.Y)
        
### Pour comparer les résultats du post-élagage et du pré-élagage, nous pouvons évaluer l'exactitude et la complexité des deux méthodes.

## En termes de précision, nous pouvons comparer les performances du meilleur arbre élagué sélectionné par post-lagage et de l'arbre élagué obtenu par pré-lagage sur l'ensemble de validation.
# Calcule des performances sur l'ensemble de validation de l'arbre élagué obtenu par pré-élagage
print("\nLa performance sur le jeu de validation pour l'arbre pré-élagué est : ", tr_pruned_pre.score(data_valid.iloc[:,:2], data_valid.Y))
# Calcule des performances sur l'ensemble de validation pour le meilleur arbre élagué sélectionné par post-élagage
print("La performance sur le jeu de validation pour l'arbre post-élagué est : ", tr_pruned_post.score(data_valid.iloc[:,:2], data_valid.Y))

## En termes de complexité, nous pouvons comparer le nombre de nœuds dans les deux arbres.
# Calculer du nombre de nœuds dans l'arbre élagué obtenu par pré-élagage
print("\nNombre de nœuds dans l'arbre élagué obtenus par pré-élagage : ", tr_pruned_pre.tree_.node_count)
# Calculer le nombre de nœuds dans l'arbre le mieux taillé sélectionné par post-élagage
print("Nombre de nœuds dans l'arbre élagué obtenus par post-élagage :", tr_pruned_post.tree_.node_count)

print("\nL'arbre pré-élagage a",tr_pruned_pre.tree_.node_count,"Nœuds et",tr_pruned_pre.tree_.n_leaves,"Feuilles.")
print("L'arbre post-élagage a",tr_pruned_post.tree_.node_count,"Nœuds et",tr_pruned_post.tree_.n_leaves,"Feuilles.")
```

Cependant, la performance de l'arbre post-élagué est meilleure que celle de l'arbre pré-élagué, ce qui suggère que l'arbre post-élagué est plus généralisable et a une capacité de prédiction plus élevée.

    En résumé, il est important d'utiliser des techniques d'élagage pour améliorer la performance et la généralisation des modèles d'arbre de décision, et le choix de la méthode d'élagage dépend des besoins spécifiques de l'application, de la taille et de la complexité du jeu de données.

## <font color="#9400D3">2. Using HOG representation rather than pixel values</font>

In the previous section, we have directy used the pixel values as features to perform classification. For complex images, the pixel values might not be adapted to this task. Better image representations can be used in order to transform an image into a feature vector that contains informations about then content of an image (colors, orientations for instance). 

We will use a quite simple image representation to perform classification : the HOG representation (some informations can be found here : 
https://scikit-image.org/docs/dev/auto_examples/features_detection/plot_hog.html).

This representation computes histograms of gradient orientations. Actually, it splits the image into sub-images (blocks) and computes one histogram per block. The number of orientations and blocks are parameters of the HOG representation. 
For instance, with our images ($28\times28$), if we ask for blocks of $14\times14$ pixels and 8 different orientations (angles), it will produce 4 (4 blocks of size  $14\times14$ in the image) histograms of size 8 (8 orientations), hence a vector of size 32.
Each image will now be represented with this vector of size 32 (rather than the 784 pixel values).
Below, an example of how to produce this vector from an image

```python
%pip install scikit-image
```

```python
from skimage.feature import hog

im= mnist.iloc[0,1:] # first image of the mnist dataset (784 pixel values)
im_hog = hog(im.to_numpy().reshape(28,28,1), orientations=8, pixels_per_cell=(14,14), cells_per_block=(1,1),multichannel=True)
print(im_hog)
```

You can see that the result is a vector of size 32 (4 blocks, and histograms with 8 orientations). Each histogram is normalized so that the sum of square values in the histogram equals 1. 

In the following, we will transform each image of our dataset using this representation and then use this new feature as input to classifiers, hoping that it will improve the classification performance.
The next two cells transform the training, validation and test set with the HOG representation (blocks of 14*14 pixels and 8 orientations, you can change these parameters later)

```python
def compute_hog(row, ori, cell):
    return(pd.Series(hog(row.iloc[1:].to_numpy().reshape(28,28,1), orientations=ori, pixels_per_cell=(cell, cell), cells_per_block=(1,1),multichannel=True)))
```

```python
hog_train = mnist_train.apply(compute_hog, axis=1, args=(8,14))
hog_valid= mnist_valid.apply(compute_hog, axis=1, args=(8,14))
hog_test = mnist_test.apply(compute_hog, axis=1, args=(8,14))
hog_train['label'] = mnist_train.label
hog_valid['label'] = mnist_valid.label
hog_test['label'] = mnist_test.label

hog_train
# you see here that each image of the training set is now a vector of length 32.
# be careful: label is now the last column
```
